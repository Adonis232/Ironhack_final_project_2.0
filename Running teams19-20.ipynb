```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
data = pd.read_csv('./data/season19-20.csv')
#data = pd.read_csv('./data/season18-19.csv')
#data = pd.read_csv('./data/season17-18.csv')
```

```python
data.head()
```

```python
# I need the data to be a cumulative list of totals for each team over the
#season
```

```python
teams = data.HomeTeam.unique()


teamlst = {Team : [0] for Team in teams}
```

```python
teamlst
```

```python
#for each row in dataframe

for row in data.itertuples():
    
    #adding home and away team names to the variable
    
    
    home = row.HomeTeam
    away = row.AwayTeam
    #giving the points to the correct team after reading the games
    if row.FTHG > row.FTAG:
        teamlst[home].append(teamlst[home][-1]+3)
        teamlst[away].append(teamlst[away][-1]+0)
        
    elif row.FTHG < row.FTAG:
        teamlst[home].append(teamlst[home][-1]+0)
        teamlst[away].append(teamlst[home][-1]+3)
        
    else:
        teamlst[home].append(teamlst[home][-1]+1)
        teamlst[away].append(teamlst[home][-1]+1)
    
```

```python
#teamlst['Wolves']
```

```python
##Thats a trial

match = list(range(0, 39))

plt.plot(match, teamlst['Southampton'], color='r')
plt.plot(match, teamlst['Everton'])
```
